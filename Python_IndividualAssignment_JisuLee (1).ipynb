```python
#  Bike Rental Analysis

# ----------------------------
# 1. Import Libraries
# ----------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")


# ----------------------------
# 2. Load the Dataset
# ----------------------------
# storing the path to the dataset
file = "bikes_data.xlsx"


# reading the file into Python through pandas
bikes_df = pd.read_excel (io         =  file,
                          sheet_name = 'bike',
                          header     =  0)

```

<hr style="height:.9px;border-width:0;color:gray;background-color:gray"><br>
<strong> <font size="3">When visualizing data trends over time, it is important to explicitly define the order of time-related variables so that the graph is correctly sorted. For this reason, I structured the code to ensure that the time, month, and season data appear in the appropriate chronological order.</font></strong>

```python
# ----------------------------
# 3. Clean and Prepare Data : Convert categorical columns
# ----------------------------

month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
season_order = ['Winter', 'Spring', 'Summer', 'Autumn']

bikes_df['Month'] = pd.Categorical(bikes_df['Month'], categories=month_order, ordered=True)
bikes_df['Day of Week'] = pd.Categorical(bikes_df['Day of Week'], categories=day_order, ordered=True)
bikes_df['Season'] = pd.Categorical(bikes_df['Season'], categories=season_order, ordered=True)
bikes_df['Holiday'] = bikes_df['Holiday'].astype('category')
```

<hr style="height:.9px;border-width:0;color:gray;background-color:gray"><br>
<strong> <font size="3">To better understand not just the time trend but also potential patterns by time category, I grouped the hours into four ranges: 'Morning', 'Afternoon', 'Evening', and 'Night'.</font></strong>

```python
# Time Catetorize
def time_of_day(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

    

bikes_df['Time_of_Day'] = bikes_df['Hour'].apply(time_of_day)
bikes_df['Time_of_Day'] = pd.Categorical(bikes_df['Time_of_Day'], 
                                         categories=['Morning', 'Afternoon', 'Evening', 'Night'], 
                                         ordered=True)
```

<hr style="height:.9px;border-width:0;color:gray;background-color:gray"><br>
<strong> <font size="3">To better understand not just the time trend but also potential patterns by time category, I grouped the hours into four ranges: 'Morning', 'Afternoon', 'Evening', and 'Night'.</font></strong>

```python
# Temperature Binning
def temp_category(temp):
    if temp < 0:
        return 'Freezing'
    elif 0 <= temp < 10:
        return 'Cold'
    elif 10 <= temp < 20:
        return 'Mild'
    elif 20 <= temp < 30:
        return 'Warm'
    else:
        return 'Hot'

bikes_df['Temperature_Bin'] = bikes_df['Temperature'].apply(temp_category)
bikes_df['Temperature_Bin'] = pd.Categorical(bikes_df['Temperature_Bin'],
                                             categories=['Freezing', 'Cold', 'Mild', 'Warm', 'Hot'],
                                             ordered=True)
```

<hr style="height:.9px;border-width:0;color:gray;background-color:gray"><br>
<strong> <font size="3"> By categorizing hours into time segments, I was able to group rental activity and analyze usage patterns more effectively throughout the day. </font></strong>

<strong> <font size="3">Graph Explanation</font></strong><br>
This boxplot illustrates the distribution of bike rentals across different times of the day. Evening hours clearly show the highest median rental count and the widest variation, indicating that this is when users are most active. Afternoon usage is also relatively consistent, while morning and night hours show lower median values with noticeable outliers. These outliers may reflect occasional surges in rental activity, potentially related to specific events or commuting behavior. This visualization helped clarify when demand is strongest throughout the day, and also pointed to quieter time slots that could benefit from targeted engagement strategies.

```python
# ----------------------------
# 4. Visualization 1: Rentals by Time of Day
# ----------------------------
plt.figure(figsize=(8,5))
sns.boxplot(x='Time_of_Day', y='RENTALS', data=bikes_df)
plt.title('Bike Rentals by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Rental Count')
plt.tight_layout()
plt.show()

```

```python
# ----------------------------
# 4. Visualization 7 : Temperature vs Rentals
# ----------------------------

# lmplot with existing Temperature_Bin
sns.lmplot(
    x='Temperature',
    y='RENTALS',
    hue='Temperature_Bin',  
    data=bikes_df,
    fit_reg=True,
    scatter=True,
    aspect=2,
    legend=False
)

# Draw vertical lines at temperature boundaries
for temp_cut in [0, 10, 20, 30]:
    plt.axvline(x=temp_cut, color='gray', linestyle='--')

    
# Format plot
plt.title('Temperature and Bike Rentals by Temperature Category')
plt.xlabel('Temperature (C)')
plt.ylabel('Number of Rentals')
plt.xlim(left=bikes_df['Temperature'].min() - 1, right=bikes_df['Temperature'].max() + 1)
plt.tight_layout()
plt.show()
```
